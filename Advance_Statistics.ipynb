```python
2. What are the types of random variables?
#  -> 1. Discrete Random Variable
   * Definition:
      A random variable that takes on a finite or countably infinite set of distinct values.
  *  Values: Typically integers.
  * Probability is described by: A Probability Mass Function (PMF).
  # Examples:
   1. Number of heads in 5 coin tosses (0, 1, 2, 3, 4, 5).
   2. Number of students present in a class.
   3. Rolling a die (1, 2, 3, 4, 5, 6).

# 2. Continuous Random Variable
  * Definition: A random variable that can take on any value within an interval (finite or infinite).
  * Values: Uncountably infinite (real numbers).
  * Probability is described by: A Probability Density Function (PDF).
  * Probability of any exact value: Always zero; only ranges have non-zero probabilities
#  Examples:
    Temperature on a given day (e.g., 23.4°C).
    Time taken to complete an exam.
    Height of students in a class.
```

```python
13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  -> states that the distribution of sample means will approximate a normal distribution, regardless
     of the original population's distribution, as the sample size becomes sufficiently large.
    #  Key aspects of the CLT:
Sample Means:-
The theorem focuses on the distribution of sample means, not individual data points.
Large Sample Size:-
      The approximation to a normal distribution improves as the sample size increases. A common rule of thumb
       is that a sample size of 30 or more is generally considered large enough.
Independence and Identical Distribution:-
The samples should be independent and identically distributed (i.i.d.) for the CLT to hold strictly.
#  is the CLT critical in statistics
* Foundation for Inference:-
Many statistical methods, like hypothesis testing and confidence interval construction, rely on the assumption of a normal distribution.
* Simplifies Analysis:-
By allowing the use of normal distribution-based techniques, the CLT simplifies the analysis of data from various distributions.
```

```python
14. How does the Central Limit Theorem relate to the normal distribution?
 -> The distribution of sample means will approximate a normal distribution, regardless of the
   underlying population distribution, as long as the sample size is sufficiently large.
  #  Normal Distribution:
   The normal distribution, also known as the Gaussian distribution, is a bell-shaped probability distribution characterized by its mean and standard deviation.
#  Central Limit Theorem:
  The CLT focuses on the distribution of sample means. It states that if you take many random samples from a population and calculate the mean of each sample, the distribution of these sample means will be approximately normal, even if the original population isn't normally distributed.
#  Relationship:
  The CLT's significance lies in its connection to the normal distribution. It allows us to use the properties of
   the normal distribution (like probabilities associated with specific ranges) to analyze and make inferences about sample means,
    even when the population distribution is unknown or non-normal.
```

```python
15. What is the application of Z statistics in hypothesis testing?
  ->  to determine if there's a significant difference between a sample mean and a population mean, or between the means of two
      samples, when certain conditions are met.

# 1. When to use z-tests:
Large sample sizes:-
Z-tests are most reliable when dealing with larger sample sizes, as the sampling distribution of the sample mean approaches a normal distribution (Central Limit Theorem).
Known population variance:-
If the population standard deviation (and thus the variance) is known, a z-test can be used.
Comparing sample means to a population mean:-
Z-tests can determine if a sample mean is significantly different from a known population mean.
# 2. Steps in using z-tests for hypothesis testing:

State the null and alternative hypotheses:-
     Define what you're trying to prove or disprove. The null hypothesis (H0) is a statement of no effect
    or no difference, while the alternative hypothesis (Ha) is the statement you're trying to find evidence for.
Set the significance level (alpha):-
This determines the threshold for rejecting the null hypothesis. Common values are 0.05 or 0.01.
```

```python
 16. How do you calculate a Z-score, and what does it represent?
-  -> subtracting the mean from the data point and dividing by the standard deviation.
  #  Formula:
z = (x - μ) / σ
  # Where:
   z: is the z-score
   x: is the individual data point
   μ: (mu) is the mean of the dataset
   σ: (sigma) is the standard deviation of the dataset
#  Meaning:
Standard Deviation:
The z-score essentially standardizes a raw score, expressing it in terms of standard deviations. A z-score of 1 means the data point is one standard deviation above the mean, while a z-score of -2 means it's two standard deviations below the mean.
#  Comparison:
Z-scores allow for comparing data points from different distributions, as they are standardized.
  # Outlier Detection:
Z-scores can help identify outliers. Data points with very high or very low z-scores (e.g., beyond +2 or -2) may be considered unusual or outliers.
#  Probability:
Z-scores are used in conjunction with the standard normal distribution to determine the probability of observing a value greater than or less than a given data point.
```

```python
17.  What are point estimates and interval estimates in statistics?
#  -> Point Estimates:
   * A point estimate is a single value calculated from sample data that is used to estimate a population parameter.
   * For example, the sample mean is a point estimate of the population mean.
   * Point estimates are straightforward to calculate and understand, but they do not provide any information about the precision or reliability of the estimate.
   * A point estimate alone doesn't tell you how close it is to the true population value.
  # Interval Estimates:
   * An interval estimate provides a range of values within which the population parameter is likely to fall, along with a specified level of confidence.
   * A common type of interval estimate is the confidence interval, which gives a range of values that is likely to contain the true population parameter.
   * Interval estimates are more informative than point estimates because they provide information about the uncertainty associated with the estimate.
   * For instance, a 95% confidence interval means that if you were to repeat the sampling process many times,
     95% of the calculated confidence intervals would contain the true population parameter
```

```python
18.  What is the significance of confidence intervals in statistical analysis?
  # -> 1. Quantifying Uncertainty:
 * Confidence intervals express the uncertainty associated with an estimate derived from a sample.
 * Instead of just stating a single value (e.g., the average income of a group), a confidence interval
    provides a range within which the true population value is likely to fall.
# 2. Reliability and Precision:
 * Wider confidence intervals indicate greater uncertainty and less precision in the estimate.
 * Conversely, narrower confidence intervals suggest a more precise estimate with less variability.
# 3. Making Data-Driven Decisions:
 * Confidence intervals enable more informed decision-making by providing a range of plausible values rather than a single point.
 * For instance, in drug trials, confidence intervals around the effect size of a new medication can help determine if it's significantly better than an existing treatment.
# 4. Complementing Significance Tests:
 * Confidence intervals offer a more complete picture than significance tests (p-values) alone.
 * While a p-value indicates whether a result is statistically significant, a confidence interval shows the magnitude and precision of the effect.
# 5. Inferences about the Population:
  * Confidence intervals are essential for making inferences about a population based on a sample.
  * By calculating a confidence interval for a sample statistic, researchers can estimate the likely range of the corresponding population paramet
```

```python
19. What is the relationship between a Z-score and a confidence interval?
  ->  the Z-score is used to calculate the margin of error, which is a key component in determining the
      width of a confidence interval.
      # Z-score:
A Z-score indicates how many standard deviations a data point is away from the mean in a standard normal
 distribution. It's often used with confidence intervals to represent how confident we are that the true population
  parameter lies within a certain range.
# Confidence Interval:
A confidence interval is a range of values calculated from sample data that is likely to contain the true
 population parameter. The confidence level associated with the interval indicates how sure we are that the
  true value falls within that range.
# Relationship:
The Z-score is directly used in the calculation of the confidence interval. Specifically, the Z-score corresponding
 to the chosen confidence level (e.g., 1.96 for 95% confidence) is used to calculate the margin of error, which is then
 added and subtracted from the sample mean to create the confidence interval.
```

```python
20.  How are Z-scores used to compare different distributions?
  # ->  1. Standardization:
Z-scores transform raw data into a standard scale where the mean is 0 and the standard deviation is
1. This is done by calculating: z = (x - μ) / σ, where x is the data point, μ is the mean of the distribution, and σ is the standard deviation.
2. Comparison:
   By converting data points to z-scores, you can compare values from different distributions on a common scale. For
    example, a z-score of +2 in one distribution means the same thing as a z-score of +2 in another distribution:
    it represents a value that is two standard deviations above the mean of its respective distribution.
3. Meaningful Interpretation:
   Z-scores indicate the relative position of a data point within its distribution. A positive z-score indicates a value
 above the mean, while a negative z-score indicates a value below the mean. The larger the absolute value of the z-score,
  the more extreme the data point is relative to its distribution.
```

```python
21. What are the assumptions for applying the Central Limit Theorem?
  -> random sampling, independence of samples, a sufficiently large sample size, and sometimes, that the sample
     size is less than 10% of the population when sampling without replacement.
    #  Random Sampling:
The sample must be drawn randomly from the population. This ensures that each member of the population
 has an equal chance of being selected, which is crucial for reducing bias and making the sample representative of the population.
#  Independence of Samples:
The individual observations within the sample should be independent of each other. This means
 that the selection of one data point should not influence the selection or value of any other data point in the sample.
#  Sufficiently Large Sample Size:
The sample size needs to be large enough for the CLT to be effective. While a rule of thumb is
 often cited as n ≥ 30, the necessary sample size can vary depending on the shape of the population
  distribution. For distributions that are heavily skewed or have outliers, a larger sample size might be needed.
#  Sampling Without Replacement:
If sampling is done without replacement, the sample size should not exceed 10% of the total population
```

```python
2.  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the discrete uniform distribution
# For example, a fair six-sided die roll (values from 1 to 6)
low_bound = 1
high_bound = 7  # Upper bound is exclusive in scipy.stats.randint

# Generate the possible outcomes (x-axis for the PMF plot)
outcomes = np.arange(low_bound, high_bound)

# Calculate the Probability Mass Function (PMF)
# For a discrete uniform distribution, each outcome has an equal probability
# which is 1 / (number of outcomes).
pmf_values = randint.pmf(outcomes, low=low_bound, high=high_bound)

# Plotting the PMF
plt.figure(figsize=(8, 6))
plt.plot(outcomes, pmf_values, 'bo', ms=8, label='PMF values') # Plot points
plt.vlines(outcomes, 0, pmf_values, colors='b', lw=3, alpha=0.7, label='PMF bars') # Plot vertical lines
plt.xlabel('Outcome (X)')
plt.ylabel('Probability P(X=x)')
plt.title(f'Probability Mass Function of a Discrete Uniform Distribution (Range: {low_bound} to {high_bound-1})')
plt.xticks(outcomes) # Ensure ticks are at each outcome
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution for a given x and probability p.

    Parameters:
    x (int): The value (0 or 1)
    p (float): The probability of success (0 <= p <= 1)

    Returns:
    float: The probability P(X=x)
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0.0  # Outside the support of Bernoulli distribution

# Example usage
print("P(X=1) with p=0.7:", bernoulli_pdf(1, 0.7))
print("P(X=0) with p=0.7:", bernoulli_pdf(0, 0.7))
print("P(X=2) with p=0.7:", bernoulli_pdf(2, 0.7))  # Invalid input
```

```python
6.  Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution>

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define parameters for the discrete uniform distribution
low = 1   # inclusive
high = 7  # exclusive; values will be from 1 to 6

# Create the discrete uniform distribution
dist = randint(low, high)

# Define range of values (support)
x = np.arange(low, high)
cdf_values = dist.cdf(x)

# Plot CDF
plt.step(x, cdf_values, where='post', label='CDF')
plt.title('CDF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.xticks(x)
plt.yticks(np.linspace(0, 1, high - low + 1))
plt.legend()
plt.show()
```

```python
9. Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_z_scores(data):
    """
    Calculates Z-scores for a given dataset.

    Parameters:
        data (array-like): List or NumPy array of numerical values.

    Returns:
        z_scores (numpy.ndarray): Z-scores of the dataset.
    """
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

# Example usage:
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=1000)  # Sample dataset
z_scores = calculate_z_scores(data)

# Plotting Z-scores
plt.figure(figsize=(10, 5))
sns.histplot(z_scores, bins=30, kde=True, stat='density', color='cornflowerblue', edgecolor='black')
plt.title('Z-score Distribution')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.axvline(x=0, color='red', linestyle='--', label='Mean (Z=0)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
original_dist = np.random.exponential(scale=2.0, size=100000)  # Non-normal distribution
sample_size = 30      # n >= 30 for CLT to hold
num_samples = 1000    # Number of samples to draw

# Collect sample means
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(original_dist, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

# Plot original non-normal distribution
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.histplot(original_dist, bins=50, kde=True, color='tomato', stat='density')
plt.title("Original Exponential Distribution")
plt.xlabel("Value")
plt.ylabel("Density")

# Plot distribution of sample means (CLT result)
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=30, kde=True, color='steelblue', stat='density')
plt.title(f"Sampling Distribution of Sample Means (n={sample_size})")
plt.xlabel("Sample Mean")
plt.ylabel("Density")

plt.tight_layout()
plt.show()
```

```python
15. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(0)

# Parameters
mu = 50                # Population mean
sigma = 10             # Population standard deviation
sample_size = 30       # Size of each sample
num_samples = 1000     # Number of samples to draw

# Step 1: Simulate multiple samples from a normal distribution
sample_means = []
for _ in range(num_samples):
    sample = np.random.normal(loc=mu, scale=sigma, size=sample_size)
    sample_means.append(np.mean(sample))

# Step 2: Plot the distribution of sample means
plt.figure(figsize=(10, 5))
sns.histplot(sample_means, bins=30, kde=True, stat='density', color='mediumseagreen', edgecolor='black')
plt.title(f'CLT Verification: Distribution of Sample Means (n={sample_size})')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.axvline(np.mean(sample_means), color='red', linestyle='--', label=f'Mean ≈ {np.mean(sample_means):.2f}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
16. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():
    """
    Calculates and plots the standard normal distribution PDF.
    """
    # Generate x values from -4 to 4
    x = np.linspace(-4, 4, 1000)

    # Calculate the PDF using scipy.stats.norm
    pdf = norm.pdf(x, loc=0, scale=1)

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.plot(x, pdf, color='blue', lw=2, label='Standard Normal PDF')
    plt.fill_between(x, pdf, alpha=0.3, color='lightblue')
    plt.title('Standard Normal Distribution (μ = 0, σ = 1)')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.axvline(0, color='black', linestyle='--', linewidth=1)

    # Add standard deviation markers
    for i in [-1, 1, -2, 2, -3, 3]:
        plt.axvline(i, color='gray', linestyle='--', alpha=0.5)
        plt.text(i, 0.01, f'{i}σ', ha='center')

    plt.grid(True)
    plt.legend()
    plt.show()

# Call the function
plot_standard_normal_distribution()
```

```python
18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """
    Calculates the Z-score for a given data point.
    Z = (x - mean) / std
    """
    return (x - mean) / std

def plot_standard_normal_with_z(z_score):
    """
    Plots the standard normal distribution and highlights the area corresponding to the Z-score.
    """
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, y_vals, label='Standard Normal PDF', color='blue')
    plt.fill_between(x_vals, y_vals, where=(x_vals <= z_score), color='lightgreen', alpha=0.5,
                     label=f'Area left of Z={z_score:.2f}')

    plt.axvline(z_score, color='red', linestyle='--', linewidth=2, label=f'Z = {z_score:.2f}')
    plt.title('Z-score on Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example Input
x = 72          # Data point
mean = 65       # Population mean
std_dev = 5     # Standard deviation

# Calculate Z-score
z = calculate_z_score(x, mean, std_dev)
print(f"Z-score for x = {x}: {z:.2f}")

# Plot the result on the standard normal distribution
plot_standard_normal_with_z(z)
```

```python
19. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

# Hypothesized population mean
mu_0 = 1000

# Sample data
sample_mean = 980
sample_std = 40
n = 50  # sample size

# Step 1: Calculate Z-statistic
z = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Step 2: Calculate two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))  # two-tailed

# Step 3: Conclusion
alpha = 0.05

print(f"Z-statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

```python
20.Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy import stats

# Step 1: Sample data (example)
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=25)  # small sample size (n < 30)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)
confidence_level = 0.95

# Step 3: Compute the margin of error using t-distribution
t_crit = stats.t.ppf((1 + confidence_level) / 2, df=n-1)
margin_of_error = t_crit * (sample_std / np.sqrt(n))

# Step 4: Compute the confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"{int(confidence_level*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Step 1: Generate data from a normal distribution
np.random.seed(1)  # for reproducibility
mu = 70           # true population mean
sigma = 10        # population standard deviation
n = 50            # sample size

data = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
standard_error = sample_std / np.sqrt(n)

# Step 3: Calculate 95% confidence interval using t-distribution
confidence_level = 0.95
t_crit = stats.t.ppf((1 + confidence_level) / 2, df=n - 1)
margin_of_error = t_crit * standard_error

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 4: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 5: Visualization
plt.figure(figsize=(8, 4))
plt.errorbar(x=0, y=sample_mean, yerr=margin_of_error, fmt='o', capsize=5, color='darkblue')
plt.axhline(mu, color='green', linestyle='--', label=f'True Mean = {mu}')
plt.xlim(-1, 1)
plt.title('95% Confidence Interval for the Mean')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define Poisson distribution parameter
lambda_val = 4  # average number of events

# Generate range of k values
k_values = np.arange(0, 15)

# Calculate CDF values
cdf_values = poisson.cdf(k_values, mu=lambda_val)

# Print the CDF values
for k, cdf in zip(k_values, cdf_values):
    print(f"P(X ≤ {k}) = {cdf:.4f}")

# Plot the CDF
plt.figure(figsize=(10, 5))
plt.step(k_values, cdf_values, where='post', color='darkorange', label=f'Poisson(λ={lambda_val}) CDF')
plt.scatter(k_values, cdf_values, color='blue')
plt.title(f'Cumulative Distribution Function of Poisson(λ={lambda_val})')
plt.xlabel('Number of Events (k)')
plt.ylabel('P(X ≤ k)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
25. Write a Python program to compare the standard deviations of two datasets and visualize the difference

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate two datasets
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=5, size=100)   # Dataset 1: small spread
data2 = np.random.normal(loc=50, scale=15, size=100)  # Dataset 2: wider spread

# Calculate standard deviations
std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

# Combine data for plotting
combined_data = [data1, data2]

# Plot
plt.figure(figsize=(8, 5))
sns.boxplot(data=combined_data)
plt.xticks([0, 1], ['Dataset 1', 'Dataset 2'])
plt.title('Comparison of Standard Deviations')
plt.ylabel('Value')

# Annotate standard deviations on plot
plt.text(0, np.max(data1) + 1, f'SD = {std1:.2f}', ha='center', color='blue')
plt.text(1, np.max(data2) + 1, f'SD = {std2:.2f}', ha='center', color='blue')

plt.grid(True)
plt.show()
```

```python
26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np

# Step 1: Generate data from a normal distribution
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=1000)

# Step 2: Calculate Range
data_range = np.max(data) - np.min(data)

# Step 3: Calculate IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Output results
print(f"Range: {data_range:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
print(f"IQR (Interquartile Range): {iqr:.2f}")
```

```python
 27. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate dataset (e.g., normal distribution)
np.random.seed(1)
data = np.random.normal(loc=50, scale=10, size=1000)

# Step 2: Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Step 3: Plot original vs normalized
plt.figure(figsize=(12, 5))

# Original data
plt.subplot(1, 2, 1)
sns.histplot(data, bins=30, kde=True, color='skyblue', edgecolor='black')
plt.title('Original Data Distribution')
plt.xlabel('Original Values')
plt.ylabel('Frequency')

# Normalized data
plt.subplot(1, 2, 2)
sns.histplot(z_scores, bins=30, kde=True, color='salmon', edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-scores')
plt.ylabel('Frequency')
plt.axvline(0, color='black', linestyle='--', label='Mean (0)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
28.  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution(n=1000, mean=0, std=1, verbose=True):
    """
    Generates normal data and calculates skewness and kurtosis.

    Parameters:
        n (int): number of samples
        mean (float): mean of the normal distribution
        std (float): standard deviation
        verbose (bool): whether to print interpretation

    Returns:
        tuple: (skewness, kurtosis)
    """
    np.random.seed(42)
    data = np.random.normal(loc=mean, scale=std, size=n)

    # Calculate skewness and kurtosis (excess by default)
    skewness = skew(data)
    kurt = kurtosis(data)  # excess kurtosis (normal = 0)

    if verbose:
        print(f"Skewness: {skewness:.4f}")
        print(f"Kurtosis (Excess): {kurt:.4f}")
        print("\nInterpretation:")
        print(" - Skewness ~ 0 indicates symmetry (normal-like).")
        print(" - Kurtosis ~ 0 indicates normal tails (mesokurtic).")

    return skewness, kurt

# Example use
analyze_distribution()
```
